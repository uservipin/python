1.In the pre-smartphone era,each number key is assigned a subset of the alphabet {a,b,â€¦,z}. 
    - 2 is assigned {a,b,c}, 
    - 3 is assigned {d,e,f} 
    - 4 is assigned {g,h,i} 
    - 5 is assigned {j,k,l}
    - 6 is assigned {m,n,o} 
    - 7 is assigned {p,q,r,s}
    - 8 is assigned {t,u,v} 
    - 9 is assigned {w,x,y,z}
    
Write a function numbers_to_chars() to find the characters generated using key 9999335533. 
Output should be "zeke"

Function Name : numbers_to_chars()
Input : Integer number sequence
Output : Str

```python
def numbers_to_chars(numbers):
    d = {"2":{'a','b','c'},"3":{'d','e','f'},"4":{'g','h','i'},"5":{'j','k','l'},"6":{'m','n','o'},"7":{'p','q','r','s'},"8":{'t','u','v'},"9":{'w','x','y','z'}}
    d2 = {}
    l = []
    current = numbers[0]
    count = 0
    output = ""
    
    for value in d:
        d2[value] = list(sorted(d[value]))

    for num in numbers:
        if num == current:
            count += 1
        else:
            l.append((current, count))
            current = num
            count = 1
    l.append((current, count))

    for t in l:
        key = d2[t[0]]
        res = key[int(t[1]-1)]
        output += res
        
    return output


# numbers_to_chars("9999335533")
```

2.Given a dict of tickets("to":"from")

{"Chennai":"Banglore","Bombay":"Delhi","Goa":"Chennai","Delhi":"Goa"} find out the sequence of travel.

Expected Output : Bombay->Delhi, Delhi->Goa, Goa->Chennai, Chennai->Banglore

Function Name : travel_sequence
Input : dict 
Output : dict

```python
def travel_sequence(d):
    l = []
    d2 = {}

    for key,value in d.items():
        temp = []
        if key not in d.values():
            temp.append(key)
            temp.append(value)
            l.append(temp)

    while len(l) < len(d):
        for key,value in d.items():
            temp2 = []
            if key == l[-1][-1]:
                temp2.append(key)
                temp2.append(value)
                l.append(temp2)

    for value in l:
        d2.setdefault(value[0],value[1])
        
    return d2


# travel_sequence({"Chennai":"Banglore","Bombay":"Delhi","Goa":"Chennai","Delhi":"Goa"})
```

3.Given a dictionary that associates the names of states with a list of the names of cities that appear in it,write a program that creates a new dictionary that associates the name of a city with the list of states that it appears in.

As an example, if the first dictionary is

Input :
states = {'New Hampshire': ['Concord', 'Hanover'],

'Massachusetts': ['Boston', 'Concord', 'Springfield'],

'Illinois': ['Chicago', 'Springfield', 'Peoria']
}

Output:

cities = {'Hanover': ['New Hampshire'],

'Chicago': ['Illinois'],'Boston': ['Massachusetts'], 

'Peoria': ['Illinois'],'Concord': ['New Hampshire','Massachusetts'],

'Springfield': ['Massachusetts', 'Illinois']
}

Function Name : city_with_states
Input : dict 
Output : dict

```python
def city_with_states(states):

    l = []
    l2 = []
    cities = {}

    for value in states:
        for value2 in states[value]:
            if value2 not in l:
                l.append(value2)

    for city in l:
        temp = []
        for state in states:
            if city in states[state]:
                temp.append(state)
        l2.append(temp)

    for pair in zip(l,l2):
        cities[pair[0]] = pair[1]

    return cities


# city_with_states({'New Hampshire': ['Concord', 'Hanover'],'Massachusetts': ['Boston', 'Concord', 'Springfield'],
#           'Illinois': ['Chicago', 'Springfield', 'Peoria'] })
```

4.How do you check if a given String contains valid parentheses?
Given a string containing just the characters '(', ')', '{', '}', '[' and ']', write a program in python to check if the input string is valid. The brackets must close in the correct order, "()" and "()[]{}" are all valid but "(]" and "([)]" are not.

Function Name : check_parentheses
Input : str
Output : True/False

```python
def check_parentheses(string):

    condition1 = True

    if string.count("(") != string.count(")"):
        condition1 = False

    if string.count("[") != string.count("]"):
        condition1 = False

    if string.count("{") != string.count("}"):
        condition1 = False

    condition2 = True

    if condition1 == True and string[0] not in ")]}" and string[-1] not in "([{":
        for index,value in enumerate(string):
            if value == "(" and string[index+1] in "]}":
                condition2 = False
                break

            if value == "[" and string[index+1] in ")}":
                condition2 = False
                break

            if value == "{" and string[index+1] in "])":
                condition2 = False
                break
    else:
        condition2 = False

    return condition2
    
    
# check_parentheses("()[]([{}])")
```

5.Write a program to convert Integer to Roman String. For example, if a given integer is 5 then your program should print "V".

Function Name : int_roman
Input : int
Output : str

```python
def int_roman(number):
    d = {"1":"I","2":"II","3":"III","4":"IV","5":"V","6":"VI","7":"VII","8":"VIII","9":"IX","10":"X"}
    number = str(number)
    roman_string = ""
    
    for value in number:
        roman_string += "'" + d[value] + "'"
        
    return roman_string


# int_roman(123)
```

7.Write a program to check the strength of a supplied password

The length of the password must be at least 8 characters in length
The password must contain at least 1 capital letter
The password must contain at least 1 digit
The password must contain at least 1 special character and allowed special characters are (!,@,#,$,&)
                                                                                           
We need to provide feedback to the user about the strength of their password

Provide the user with a list of reasons why their password is 'weak'                                                                                       
Function Name : check_password_strength
Input : str
Output : tuple (str,list) eg ("Weak",["The password must contain at least 1 capital letter"])                                                                                          

```python
def check_password_strength(password):

    condition1 = [True, "Password must be of at least 8 characters"]

    if len(password) < 8:
        condition1[0] = False

    condition2 = [False, "The password must contain at least 1 capital letter"]

    for char in password:
        if char.isupper():
            condition2[0] = True
            break

    condition3 = [False, "The password must contain at least 1 digit"]

    for char2 in password:
        if char2.isdigit():
            condition3[0] = True
            break

    condition4 = [False, "The password must contain at least 1 special character"]

    for char3 in password:
        if char3 in "!@#$&":
            condition4[0] = True
            break

    conditions = [condition1, condition2, condition3, condition4]
    reasons = []

    for value in conditions:
        if value[0] == False:
            reasons.append(value[1])

    for value2 in conditions:
        if value2[0] == False:
            return ("Weak",reasons)
            break
    else:
        return ("Strong",reasons)
        
        
# check_password_strength("passWord1")
```

8. Check if given sentence is syntactically correct or not

A simple sentence if syntactically correct if it fulfills given rules. The following are given rules.
 - Sentence must start with a Uppercase character (e.g. Noun/ I/ We/ He etc.)
 - There must be spaces between words.
 - Then the sentence must end with a full stop(.).
 - Two continuous spaces are not allowed.
 - Two continuous uppercase characters are not allowed.
 - However the sentence can end after an uppercase character.
 
Function Name : check_sentence
Input : str
Output : tuple (True/False,list) eg (False,["There must be spaces between words."]) 

```python
import re

def check_sentence(sentence):

    condition1 = [False, "Sentence must start with a Uppercase character"]

    r1 = re.compile("^[A-Z]")
    m1 = re.findall(r1,sentence)

    if len(m1) > 0:
        condition1[0] = True

    condition2 = [False, "There must be spaces between words"]

    r2 = re.compile("[a-zA-Z]{1,12}[\s][a-zA-Z]{1,12}")
    m2 = re.findall(r2,sentence)

    if len(m2) > 0:
        condition2[0] = True

    condition3 = [False, "The sentence must end with a full stop or an upper case alphabet"]

    r3 = re.compile("[\.A-Z]$")
    m3 = re.findall(r3,sentence)

    if len(m3) > 0:
        condition3[0] = True

    condition4 = [False, "Two continuous spaces are not allowed"]

    r4 = re.compile("[\s][\s]")
    m4 = re.findall(r4,sentence)

    if len(m4) == 0:
        condition4[0] = True

    condition5 = [False, "Two continuous uppercase characters are not allowed"]

    r5 = re.compile("[A-Z][A-Z]")
    m5 = re.findall(r5,sentence)

    if len(m5) == 0:
        condition5[0] = True

    conditions = [condition1, condition2, condition3, condition4, condition5]
    reasons = []

    for value in conditions:
        if value[0] == False:
            reasons.append(value[1])

    for value2 in conditions:
        if value2[0] == False:
            return (False,reasons)
            break
    else:
        return (True,reasons)


# check_sentence("Here's an example sentence.")
```

9.Given an array arr[] of integers and an integer K, the task is to find the greatest contiguous sub-array of size K. 

Sub-array X is said to be greater than sub-array Y if the first non-matching element in both the sub-arrays has a greater value in X than in Y.

For example : 
Input: arr[] = {1, 4, 3, 2, 5}, K = 4
Output: 4 3 2 5
Two subarrays are {1, 4, 3, 2} and {4, 3, 2, 5}. First non-matching element from array1 and array 2 : 1 and 4 as 4 is greater 
Hence, the greater one is {4, 3, 2, 5}

Function Name : greatest_sub_array()
Input : list
Output : list

```python
def greatest_sub_array(arr):

    K = 4
    l = []

    for index,value in enumerate(arr):
        sub = arr[index:index+K]
        if len(sub) == K:
            l.append(sub)

    for value in zip(l[0],l[1]):
        if value[0] != value [1]:
            non_matching = value
            break

    if non_matching[0] > non_matching[1]:
        return l[0]
    else:
        return l[1]
        
        
# greatest_sub_array([1,4,3,2,5])
```

10.Given a list of N integers. The task is to eliminate the minimum number of elements such that in the resulting list the sum of any two adjacent values is even.

Numbers = [1, 3, 5, 4, 2]
Output = [1, 3, 5] 
Total elements removed 2
Elements to be removed [4,2]

Function Name : adj_sum_even()
Input : list
Output : tuple(int,list)

```python
def adj_sum_even(numbers):
    
    odd_nums = []
    even_nums = []
    total_removed = 0
    
    for value in numbers:
        if value % 2 == 0:
            even_nums.append(value)
        else:
            odd_nums.append(value)
        
    if len(odd_nums) < len(even_nums):
        for value in odd_nums:
            numbers.remove(value)
            total_removed += 1
    else:
        for value in even_nums:
            numbers.remove(value)
            total_removed += 1
            
    return (total_removed,numbers)


# adj_sum_even([1,3,5,4,2])
```

```python
def create_dataframe_of_multiple_frame(data_frame_q5c, input_data):
    data_frame_q5c.insert(loc= 0 , column= 'Respondent_Id', value=input_data['Respondent_ID'])
    data_frame_q5c.insert(loc= 1 , column= 'Industry', value=input_data['Industry'])
    data_frame_q5c.insert(loc= 2 , column= 'Size', value=input_data['Size'])
    data_frame_q5c.insert(loc= 3 , column= 'Account Manager', value=input_data['Account Manager'])
    data_frame_q5c.insert(loc= 4 , column= 'Language', value=input_data['Please select your preferred language '])
    return data_frame_q5c
def  unpivot (data_frame,Que_no,Ans):
    df_f = data_frame.melt(id_vars = ['Respondent_Id','Industry',"Size","Account Manager","Language" ], var_name =Que_no , value_name = Ans)
    return df_f

def titleToNumber(s):
	# This process is similar to binary-to-
	# decimal conversion
	result = 0;
	for B in range(len(s)):
		result *= 26;
		result += ord(s[B]) - ord('A') + 1;

	return result;
```
